# Dependencies

```python
import pandas as pd
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
```

# Cleaning Data

```python
self_df = pd.read_csv('Resources/data.csv')
self_df.head()
```

```python
#Dropping unnecessary columns
clean_df = self_df.drop(['Notes', 'Unnamed: 19'], axis = 1)
#Renaming columns
clean_df = clean_df.rename(columns = {'Energy (1-5)' : 'Energy', 'Happiness (1-5)' : 'Happiness'})
clean_df.head()
```

```python
#Segmenting dataset
base_df = clean_df[['Month', 'Day', 'Hour', 'Energy', 'Happiness']]
binary_df = clean_df[['Eat', 'Shower', 'Snack', 'Music', 'Energy', 'Happiness']]
rating_df = clean_df[['Hydrate', 'Caffeine', 'Interaction', 'Partner', 'Energy', 'Happiness']]
discrete_df = clean_df[['Activity', 'Break', 'Energy', 'Happiness']]
continuous_df = clean_df[['Temp', 'Energy', 'Happiness']]
categorical_df = clean_df[['Weather', 'Main Activity', 'Energy', 'Happiness']]
```
